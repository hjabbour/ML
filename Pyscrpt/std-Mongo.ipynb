```python
def dirext(directory:str,extension:str,filename:str)->list:
    filelist = []
    for x in os.listdir(directory):
        if x.endswith(extension):
            if x.startswith(filename):
                #print(x)
                filelist.append(x)
    return filelist



def mongoadd(files:list,symbol:str,conn):
    for file in files:
        frame=pd.read_csv(file)
        frame.drop_duplicates(subset=['Time','Indicator'])
        #frame.set_index(['Time','Indicator'],inplace=True)
        conn.append(symbol,frame)
```
