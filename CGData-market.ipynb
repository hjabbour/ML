```python
root_url = 'https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&order=market_cap_desc&per_page=1000&page=1&sparkline=false&price_change_percentage=365d'
url = root_url
data = json.loads(requests.get(url).text)
dflive = pd.DataFrame(data)
```

```python
dflive.head()
```

```python
filter = dflive.loc[:,('id','symbol','name','current_price','market_cap','fully_diluted_valuation','total_volume','high_24h','total_volume','price_change_percentage_24h','circulating_supply','total_supply','max_supply','ath','ath_change_percentage')]
filter['supply_ratio'] =dflive['circulating_supply']/filter['max_supply']
filter['cap_ratio'] =dflive['market_cap']/filter['fully_diluted_valuation']

#drop = filter.loc[(filter['supply_ratio']>0.8)&(filter['market_cap']<1000000000),:]
#drop = filter.loc[(filter['supply_ratio']>0.5)&(filter['cap_ratio'] > 0.5)&(filter['market_cap']<2000000000),:]
drop = filter.loc[(filter['cap_ratio'] > 0.5)&(filter['market_cap']<5000000000),:]

drop.sort_values(by=['supply_ratio'],ascending=False,inplace = True) 
drop 
#sort = drop.sort_values(by=['supply_ratio'],ascending=False,inplace = True) 
#filter1
#filter1 = drop.sort_values(by=['supply_ratio'],ascending=False,inplace = True) 
#filter1
#sort
```

```python
change = drop[['symbol','price_change_percentage_24h']]
change.sort_values(by=['price_change_percentage_24h'],ascending=False,inplace = True) 
change
```

```python
capratio = drop[['symbol','supply_ratio','cap_ratio']]
capratio.symbol =capratio.symbol+'usdt'
capratio.symbol =capratio.symbol.str.upper()
capratio
```
